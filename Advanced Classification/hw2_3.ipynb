```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
```

```python
with open('/content/gdrive/MyDrive/hw2/data/mnist_small.pkl', 'rb') as f:
    data = pickle.load(f)
```

```python
X = data['X']
Y = data['Y']
```

```python
Y = Y.reshape(10000)
```

```python
def plot_2d_data(X_2d, labels, title):
    plt.figure(figsize=(8, 6))
    for i in range(10):
        plt.scatter(X_2d[labels == i, 0], X_2d[labels == i, 1], label=str(i))
    plt.title(title)
    plt.xlabel('Principal Component 1' if 'PCA' in title else 't-SNE Component 1')
    plt.ylabel('Principal Component 2' if 'PCA' in title else 't-SNE Component 2')
    plt.legend()
    plt.show()
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
plot_2d_data(X_pca, Y, 'PCA Projection')
```

```python
tsne = TSNE(n_components=2)
X_tsne = tsne.fit_transform(X)
plot_2d_data(X_tsne, Y, 't-SNE Projection')
```
